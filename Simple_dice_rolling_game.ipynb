Dice Rolling Gmae

```python
import random
import time

print("ðŸŽ² Welcome to the Dice Rolling Game!")
print("You and the computer will each roll a die. Let's see who wins!")
print("----------------------------------")

player_score = 0
computer_score = 0
rounds = 5  # number of rounds

for i in range(1, rounds + 1):
    input(f"\nðŸŽ¯ Round {i}: Press Enter to roll the dice...")

    player_roll = random.randint(1, 6)
    computer_roll = random.randint(1, 6)

    print("Rolling...")
    time.sleep(1)
    print(f"You rolled: ðŸŽ² {player_roll}")
    print(f"Computer rolled: ðŸŽ² {computer_roll}")

    if player_roll > computer_roll:
        print("âœ… You win this round!")
        player_score += 1
    elif player_roll < computer_roll:
        print("ðŸ’» Computer wins this round!")
        computer_score += 1
    else:
        print("ðŸ˜ It's a tie!")

print("\n----------------------------------")
print(f"Final Score -> You: {player_score} | Computer: {computer_score}")

if player_score > computer_score:
    print("ðŸŽ‰ You won the game!")
elif player_score < computer_score:
    print("ðŸ’€ Computer won the game!")
else:
    print("ðŸ¤ It's a draw!")
```
